## Prepare Your Dataframe

Import any libraries you need and create a dataframe.

```python
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('darkgrid')
sns.set(font_scale=1.5)
import statsmodels.api as sm

df = pd.read_csv("C:\\Users\\kap19\\OneDrive\\Desktop\\LaunchCode\\Codes\\Assignments\\Assignment 4 Final Project\\Checkpoint-5-Assignment-4\\dataset.csv")
df.head()
```

```python
#Ensurring that Target column would be evaluated in heatmap
df["Target"].replace(["Dropout", "Enrolled", "Graduate"], [0,1,2], inplace=True)
df.head()
```

## Find Correlations

Use either pairplot or a heatmap or both to find the two variables with the strongest correlation in your dataset.

```python
corr = df.corr()
df_viz, ax = plt.subplots(figsize=(13,10)) 


df_viz = sns.heatmap(corr, ax=ax,
                xticklabels=corr.columns.values,
                yticklabels=corr.columns.values,
                cmap='viridis')
```

## Create Your Model

Use the two columns with the strongest correlation to create and train your model. Make sure to print out the summary and plot the column values and the line produced by the model.

```python
X = df["Marital status"]
y = df["Age at enrollment"]

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size = 0.7, 
                                                    test_size = 0.3, random_state = 100)
X_train
y_train
X_train_sm = sm.add_constant(X_train)
lr = sm.OLS(y_train, X_train_sm).fit()

lr.params
lr.summary()
```

```python
plt.scatter(X_train, y_train)
plt.plot(X_train, 15.4761 + 6.6142*X_train, 'r')
plt.show()
```

## Error Terms

Finally, plot your error terms!

```python
y_train_pred = lr.predict(X_train_sm)

res = (y_train - y_train_pred)

fig = plt.figure()
sns.distplot(res, bins = 15)
plt.title('Error Terms', fontsize = 15)
plt.xlabel('y_train - y_train_pred', fontsize = 15)
plt.show()
```
